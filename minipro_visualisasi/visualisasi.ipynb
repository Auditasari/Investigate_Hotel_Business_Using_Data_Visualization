```python
# mengisi nilai null pada kolom agent dan city dengan modus 
df['agent'].fillna(df['agent'].mode()[0], inplace = True)
df['city'].fillna(df['city'].mode()[0], inplace=True)
```

```python
# menghapus kolom company dan nilai null pada kolom children

df.drop('company', axis = 1, inplace = True)
df.dropna(subset=['children'], inplace = True)
```

```python
# mengganti tipe data pada kolom childern menjadi integer 

df['children'] = df['children'].astype(int)
df['children'].value_counts()
```
